```python
##Assignment15 ipynb
## charles wiredu
## importing the required libraries
import tweepy
import keys
from textblob import TextBlob
import preprocessor as p ## tweet-preprocessor
import pandas as pd

## configure tweet-preprocessor
p.set_options(p.OPT.URL,p.OPT.RESERVED,p.OPT.MENTION,p.OPT.NUMBER)


## set up lists to import into pandas
tweet_columns= ['handle', 'language','created at', 'app used','num words','sentiment','tweet text']
##list of items, each is a list for one tweet of data in the columns
tweet_list= []
# connect to Twitter API
client= tweepy.Client(bearer_token= keys.bearer_token)

# ask user for which handle to look at
this_user= input('Enter a twitter handle: ')
print('-'*35) # make it look nice

## search for tweets from this user, but not a retweet
this_query= f' from:{this_user} -is:retweet' ## do not include retweets
this_query= f' from:{this_user} -is:reply'
tweets= client.search_recent_tweets(query= this_query,
                                   tweet_fields=['lang','created_at','source'],
                                   max_results=50)

##print results
if tweets.data: # see if there were results returned
    for tweet in tweets.data:
        ##prepare the text
        processed_tweet= p.clean(tweet.text)
        
        # add to list, to import into Pandas later
        # ['handle', 'language','created at', 'app used', 'tweet text']
        tweet_list.append ([this_user,tweet.lang,tweet.created_at,tweet.source,len(processed_tweet.split()),TextBlob(processed_tweet).sentiment.polarity, processed_tweet])
        ##for each tweet returned, print some information
        #print(f"{'From: ':<15}  {this_user}")
        #print(f"{'Language: ':<15}  {tweet.lang}")
        #print(f"{'Created at: ':<15}  {tweet.created_at}")
        #print(f"{'App used to tweet: ':<15}  {tweet.source}")
        ##preprocess the text
        ##processed_tweet= p.clean(tweet.text)
        ##processed_tweet= tweet.text
        #print(f"{'Tweet:':<15}\n {processed_tweet}")
        #print('-'*35)
else:
    print('No recent tweets with those search parameters')

#print('Colums: ', tweet_columns)
#print(tweet_list)
#print('-'*100)
    
# create Pandas Dataframe with the tweet data(sec 7.14,9.12)
tweet_df= pd.DataFrame(tweet_list, columns= tweet_columns)
pd.options.display.max_colwidth = 80 # widest column is 80 characters wide
pd.options.display.precision = 2 # just show 2 decimal places
print(tweet_df)
print()
print('=====================================================')
print('General Statistics about the recent tweets by Skynews')
print('=====================================================')
print(tweet_df.describe())
```
